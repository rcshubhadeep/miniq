## Geometry of Complex Numbers in Python

```python
import matplotlib.pyplot as plt
import numpy as np # Not really needed at this moment
import cmath

%matplotlib inline
```

```python
c = 1+2j # Let's start with this
print("The original complex number Re{%d} Im{%d} and Length - %d" % (c.real, c.imag, abs(c)))
```

The length is defined as $ | a+ib | =  \sqrt{a^2 + b^2} $

### Cartesian representation (Argand Plane) of Complex Number

```python
fig = plt.figure(figsize=(16, 8))
origin = [0], [0]
plt.quiver(*origin, c.real, c.imag, color=['r','b','g'], scale=21)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.show()
```

### Polar representation of Complex Numbers

```python
fig = plt.figure(figsize=(16,8))
plt.polar([0, cmath.phase(c)],[0, abs(c)],marker='o')
```

The Angle for $ a+ib $ is defined as $ \theta = \tan ^{ - 1}(\frac{b}{a}) $

Using euler identity we can get rid of the trigonmetric fucntions and write it in the following way- 

$ a+ib = \sqrt{a^2 + b^2} * e ^ {i*\tan ^{ - 1}(\frac{b}{a})} $

If we express $ r = \sqrt{a^2 + b^2} $ and $ \theta = \tan ^{ - 1}(\frac{b}{a}) $ then the above can be written as 

$ a+ib = re ^ {i\theta} $

```python
# Proving euler identity in Python (fun activity)
import math
print ((math.e**(math.pi*1j)).real + 1)
```

```python
def rotate(number, degree):
    n = number
    return (math.e**(math.radians(degree)*1j)) * n
```

```python
c2 = rotate(c, 45)
```

```python
fig = plt.figure(figsize=(16,8))
cs = c, c2
for c in cs:
    plt.polar([0, cmath.phase(c)],[0, abs(c)],marker='o')
```

```python
def scale(number, scale):
    n = number
    return n * scale
```

```python
c3 = rotate(scale(1+2j, 2), 45)
print("After the operation Re{%d}, Im{%d} and legth - %d" % (c3.real, c3.imag, abs(c3)))
```

The above is the double of the first one in legth $ 4 = 2 * 2 $

```python
fig = plt.figure(figsize=(16,8))
cs = 1+2j, c3
for c in cs:
    plt.polar([0, cmath.phase(c)],[0, abs(c)],marker='o')
```
